## Q1. Load the 'titanic' dataset using seaborn and create a scatter plot for **age** and **fare** using Plotly Express.

```python
import seaborn as sns
import plotly.express as px

# Load dataset
titanic = sns.load_dataset('titanic')

# Scatter plot
fig = px.scatter(titanic, x='age', y='fare', title='Age vs Fare (Titanic Dataset)')
fig.show()
```

## Q2. Using the tips dataset in Plotly library, plot a **box plot** using Plotly Express.

```python
tips = px.data.tips()

fig = px.box(tips, x='day', y='total_bill', title='Box Plot of Total Bill by Day')
fig.show()
```

## Q3. Using the tips dataset, plot a histogram for x='sex' and y='total_bill'.
Use **smoker** column with pattern_shape parameter and **day** column with color parameter.

```python
fig = px.histogram(
    tips,
    x='sex',
    y='total_bill',
    color='day',
    pattern_shape='smoker',
    title='Histogram of Total Bill by Sex'
)

fig.show()
```

## Q4. Using the iris dataset, plot a **scatter matrix plot** using species column for color.

```python
iris = px.data.iris()

fig = px.scatter_matrix(
    iris,
    dimensions=['sepal_length','sepal_width','petal_length','petal_width'],
    color='species',
    title='Scatter Matrix of Iris Dataset'
)

fig.show()
```

## Q5. What is Distplot?
**Distplot** is used to visualize the distribution of a dataset. It combines a histogram with a density curve to show how values are distributed.

### Plot a Distplot using Plotly Express

```python
fig = px.histogram(iris, x='sepal_length', marginal='rug', title='Distribution Plot of Sepal Length')
fig.show()
```
